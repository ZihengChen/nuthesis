```python
import glob
import os, sys
import utility_common as common
from shutil import copyfile

```

```python
channels = ["ee","mumu","emu","etau","mutau","ejet","mujet"]

leptonVariables = [
    "lepton1_pt","lepton1_eta","lepton2_pt","lepton2_eta",
    "dilepton1_mass","dilepton1_pt","dilepton1_delta_r"]

for ch in channels:
    for lv in leptonVariables:
        files = glob.glob("../{}_2016/*/linear/{}/{}.png".format(ch,'lepton',lv))
        files += glob.glob("../{}_2016/*/linear/{}/{}.png".format(ch,'jet',"n_bjets"))
        files += glob.glob("../{}_2016/*/linear/{}/{}.png".format(ch,'jet',"n_jets"))
        files += glob.glob("../{}_2016/*/linear/{}/{}.png".format(ch,'misc',"met_mag"))
        for f in files:
            t = f[3:]
            t = t.replace("/","_")
            print(t)
            copyfile(f, t)
```
